Extraindo Dados da API IBGE

```python
import requests
import pandas
import json
from pyspark.sql.functions import current_timestamp, lit

API_URL = "https://servicodados.ibge.gov.br/api/v1/localidades/distritos"

def get_data(url):
    response = requests.get(url)
    return response.json()

data = get_data(API_URL)

#df_pandas to spark dataframe
df_spark = spark.createDataFrame([(json.dumps(data),)], ["raw_json"]) \
    .withColumn("ingestion_timestamp", current_timestamp()) \
    .withColumn("source", lit("ibge_api_districts"))
    
df_spark.write.mode("append").saveAsTable("project_apis.bronze.districts_ibge_raw")

```
